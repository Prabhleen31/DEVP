```python
# 9.2. Test the Significance of Correlation between 'mpg' & 'wt'.
```

```python
# Null Hyposesis : Correlation = 0 (p-value >= 5% by default)

# Correlation between Car "mpg" and "wt"
correlation_test_pearson = sps.pearsonr(car_data.mpg, car_data.wt)
correlation_test_pearson
# Inference : Car mpg & wt are negatively Correlated (p-value < 5%)
```

```python
correlation_test_spearman = sps.spearmanr(car_data.mpg, car_data.hp)
correlation_test_spearman
# Inference : Car mpg & wt are negatively Correlated (p-value < 5%)
```

```python
def vs_def(vs):
    if vs == 1:
        return("Straight")  
    else:
        return ("Vshaped")
car_data['vs_def'] = car_data['vs'].map(vs_def)
display(car_data.head(n=32))
```

```python
car_data_vshaped = car_data[car_data['vs_def'] == 'Vshaped']
car_data_vshaped [['vs']]
```

```python
car_data_straight = car_data[car_data['vs_def'] == 'Straight']
car_data_straight [['vs']]
```

```python
var_test_levene = sps.levene(car_data_vshaped.hp, car_data_straight.hp)
var_test_levene
# Inference : Variance of hp between V-Shaped & Straight Cars are (Statistically) Equal (p-value >= 5%)
```

```python
# 10.1. Build a Regression Model to Predict 'mpg' with Predictors like {hp, wt, am}. 
# 10.2. Comment on the Model Significant, the Contribution of each Predictors, & the Overall Fit.
```

```python
import statsmodels.formula.api as smf
# Model 1 : mpg = b0 + b1*disp + b2*hp + b3*wt + e
linear_regression_model1 = smf.ols('mpg ~ am + hp + wt', data=car_data).fit()
linear_regression_model1.summary()
# Inference : Model 1 Significant | hp & wt Negative & Significant, am Insignificant | Rsq (adj) = 0.84
```
