```python
import numpy as np
from numpy import linalg as npla
a = np.array([[1, 2, 1], [-1, 1, 1], [1,-3, 1]])
b = np.array([4, 0, 2]) 
npla.solve(a,b)              
```

```python
#3. Create Update & Subset Dataframe
#3.1. Create a Dataframe 'car_df' from the Dictionary 'car_dict' [05 marks]
```

```python
import pandas as pd
car_dict = {'Car': ['ABC', 'DEF', 'GHI', 'JKL', 'MNO', 'PQR', 'STU'],
    'Drive_type': ['Manual', 'Automatic', 'Automatic', 'Maunual', 'Manual','Automatic', 'Manual'],
    'car_size': ['Medium', 'Small', 'Medium','Large','Small', 'Small', 'Medium'], 
     'Horsepower': [180, 150, 250, 300, 220, 200, 180]}
car_df = pd.DataFrame(car_dict)
car_df
```

```python
#3.2. Update the Dataframe 'car_df' with the following Information: [05 marks]
```

```python
car_dict2= {"Mileage" : [18,20,14,12,24,22,16],
            "Engine_Displacement" : [1500,800,1000,2000,1000,800,1200]}
car_df2 = pd.DataFrame(car_dict2)
car_df2
```

```python
car_df = pd.concat([car_df, car_df2], axis = 1)
car_df
```

```python
#3.3. Create a Subset ‘car_df_engine_mileage’ from ‘car_df’ with the following Variables: {Car, Engine_Displacement, Mileage} [05 marks]
```

```python
car_df_engine_mileage = car_df[['Car', 'Engine_Displacement', 'Mileage']]
car_df_engine_mileage
```

```python
#4. Sort Dataframes
#4.1. Create a Copy of 'car_df' Named as 'car_df_sorted_mileage' and Sort {Highest to Lowest} by 'Mileage'
```

```python
car_df_sorted_mileage = car_df.sort_values('Mileage')
car_df_sorted_mileage
```

```python
#4.2. Create a Copy of 'car_df' Named as 'car_df_sorted_hp_mileage' and Sort: First by 'Horsepower' {Highest to Lowest}, Second by 'Mileage' {Lowest to Highest} [05 marks]
```

```python
car_df_sorted_hp_mileage = car_df.sort_values(['Horsepower', 'Mileage'], ascending=[True, False])
car_df_sorted_hp_mileage
```

```python
#5.2. Create 2 Subsets: ‘car_df_manual’ & ‘car_df_automatic’ from ‘car_df’ such that ‘car_df_manual’ & ‘car_df_automatic’ will contain all Information related Cars having ‘Drive_Type’ = ‘Manual’ & ‘Drive_Type’ = ‘Automatic’, respectively
```

```python
car_df_manual = car_df[car_df['Drive_type'] == 'Manual']
car_df_manual
```

```python
car_df_automatic = car_df[car_df['Drive_type'] == 'Automatic']
car_df_automatic
```

```python
#6. Merge Dataframe
#6.1. Inner Merge ‘car_df_manual’ having only following Variables {Car, Car_Size, Horsepower} with ‘car_df_automatic’ having only following Variables {Car, Car_Size, Mileage} 
```

```python
car_df_merged_inner = pd.merge(car_df_manual[['Car', 'car_size','Horsepower']], car_df_automatic[['Car', 'car_size','Mileage']], on='car_size', how='inner')
car_df_merged_inner
```
