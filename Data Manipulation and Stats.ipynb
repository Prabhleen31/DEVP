```python
import numpy as np
mpg_low = []
mpg_average = []
mpg_high = []
g = np.median (cars_mpg)
print ('Median value of mpg is:',g)
for i in cars_mpg:
    if i < g:
        mpg_low.append(i)
    elif i == g:
        mpg_average.append (i)
    else:
        mpg_high.append(i)
print('Values lower than median:',mpg_low)
print('Values equal to median', mpg_average)
print('Values greater than median',mpg_high)
```

```python
car_data = cars[['mpg', 'cyl', 'disp', 'hp', 'wt', 'vs', 'am']]
car_data
```

```python
mpg_mean = np.mean(car_data.mpg)
mpg_mean1 = round(mpg_mean,2)
print('Mean of the mpg values is:', mpg_mean1)
```

```python
mpg_std = np.std(car_data.mpg)
mpg_std1 = round(mpg_std,2)
print ('The std for mpg data is:',mpg_std1)
```

```python
# Cumulative Distribution Function (CDF) : Probability (p) = cdf(value, mean, sd)
```

```python
import scipy.stats as sps # Probability of a Car having mpg < 15 ## mean and std have to be defined first
p_lt15 = sps.norm.cdf(15, loc=mpg_mean, scale=mpg_std).round(2)
p_lt15
```

```python
# Probability of a Car having mpg > 25 
p_lt25 = (1- sps.norm.cdf(25, loc=mpg_mean, scale=mpg_std)) #cdf=cumulative distribution function
p_lt25_a = round (p_lt25,2)
p_lt25_a
```

```python
# Probability of a Car having 15 <= mpg <= 25
p_lt25_lt_15 = sps.norm.cdf(25, loc=mpg_mean, scale=mpg_std) - sps.norm.cdf(15,loc=mpg_mean, scale=mpg_std)
p_lt25_lt_15_a = round(p_lt25_lt_15,2)
p_lt25_lt_15_a
```

```python
# Percentage Point Function (Inverse of CDF) : Value (x) = ppf(prob, mean, sd)
```

```python
# Maximum mpg of a Car at Bottom 10% of mpg
```

```python
mpg_b10 = sps.norm.ppf(0.10, loc=mpg_mean, scale=mpg_std).round(2)
mpg_b10
```

```python
# Minimum mpg of a Car at Top 10% of mpg
```

```python
mpg_t10 = sps.norm.ppf(0.90, loc=mpg_mean, scale=mpg_std).round(2)
mpg_t10
```

```python
# mpg of a Car at Median Value of mpg
```

```python
mpg_med = sps.norm.ppf(0.5, loc=mpg_mean, scale=mpg_std).round(2)
mpg_med
```

```python
# 2. Data Analysis : Inferential Statistics
```

```python
# Shapiro-Wilk Test : Null Hypothesis :: Data is Normally Distributed (p-value >= 5% by default)

# Test of Normality of "mpg" data
normality_test_shapiro = sps.shapiro(car_data['mpg']) 
normality_test_shapiro
# Inference : "mpg" is Normally Distributed (p-value >= 5%)
```

```python
#since the p value is greater than 0.05, we accept the null hypothesis ie the data is normally distributed 
```

```python
# Comparision of "mpg" between "automatic" and "manual" Cars
```

```python
car_data_automatic = car_data[car_data.am==0]
car_data_automatic
```

```python
car_data_manual = car_data[car_data.am==1]
car_data_manual
```

```python
# TTest : Null Hypothesis :: Mean of Population 1 (Sample 1) = Mean of Population 2 (Sample 2) | (p-value >= 5% by default)
```

```python
mean_test_ttest = sps.ttest_ind(car_data_automatic.mpg, car_data_manual.mpg, equal_var=True) 
mean_test_ttest
# Inference : Average mpg of Automatic Cars < Average mpg of Manual Cars (p-value < 5%)
```

```python
# Test of Mean of Two Populations : Non Parametric Test :: Mann-Whitney Test 
# ++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++

# Null Hypothesis : Median of Population 1 (Sample 1) = Median of Population 2 (Sample 2) | (p-value >= 5% by default)

median_test_mannwhitney = sps.mannwhitneyu(car_data_automatic.mpg, car_data_manual.mpg)
median_test_mannwhitney
# Inference : Median mpg of Automatic Cars is Different from Median mpg of Manual Cars (p-value < 5%)
```

```python
import statsmodels.formula.api as smf
linear_regression_model1 = smf.ols('mpg ~ disp + hp + wt', data=car_data).fit()
linear_regression_model1.summary()
# Inference : Model 1 Significant | hp & wt Negative & Significant, disp Insignificant | Rsq (adj) = 0.81
```

```python
# --- Multiple Regression with Categorical Variables ---
# ------------------------------------------------------

# Model 2 : mpg = b0 + b1*hp + b2*wt + b3*cyl + e # Ordinal Category
linear_regression_model2 = smf.ols('mpg ~ hp + wt + C(cyl)', data=car_data).fit()
linear_regression_model2.summary()
# Inference : Model 2 Significant | hp & wt Negative & Significant; cyl 4 & 6 Negative & Significant, cyl 8 Insignificant | Rsq (adj) = 0.84
```
