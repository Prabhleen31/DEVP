```python
# 1.3. Create a Categorical Variable 'mpg_type' from 'mpg' with the following criteria {mpg_type = 'low' for Cars Below 1st Quartile of mpg Data | mpg_type = 'average' for Cars in the IQR of mpg Data | mpg_type = 'high' for Cars Above 3rd Quartile mpg Data}
```

```python
import numpy as np
mpg_low = []
mpg_average = []
mpg_high = []
g = np.median (mpg_type)
print ('Median value of mpg is:',g)
for i in mpg_type:
    if i < g:
        mpg_low.append(i)
    elif i == g:
        mpg_average.append (i)
    else:
        mpg_high.append(i)
print('Values lower than median:',mpg_low)
print('Values equal to median', mpg_average)
print('Values greater than median',mpg_high)        
```

```python
def mpg_def(mpg):
    if mpg < 19.2:
        return("Low")  
    elif mpg > 19.2:
        return ("High")
    else:
        return ("Average")
car_data['mpg_def'] = car_data['mpg'].map(mpg_def)
display(car_data.head(n=32))
```

```python
# 2.1. Create a New Variables: 'am_def' (where, am = 0 => am_def = 'automatic' & am = 1 => am_def = 'manual'}, & 'vs_def' (where, vs = 0 => vs_def = 'vshaped' & vs = 1 => vs_def = 'straight').
```

```python
p = car_data[car_data.am==1]
print('automatic',p)
q = car_data[car_data.am==0]
print('manual',q)
```

```python
def am_def(am):
    if am == 1:
        return("Automatic")  
    else:
        return ("Manual")
car_data['am_def'] = car_data['am'].map(am_def)
display(car_data.head(n=32))
def vs_def(vs):
    if vs == 1:
        return("Straight")  
    else:
        return ("Vshaped")
car_data['vs_def'] = car_data['vs'].map(vs_def)
display(car_data.head(n=32))
```

```python
# 2.2. Create a New Dataset 'cars_data' with the following variables {mpg, cyl, disp, hp, wt, vs_def, am_def}. Rename 'vs_def' & 'am_def' as 'vs' & 'am' respectively. Drop the Index. [Henceforth, Work with 'cars_data'].
```

```python
cars_data = car_data[['mpg','cyl','disp','hp','wt','am_def','vs_def']]
cars_data
```

```python
cars_data.rename(columns = {'am_def':'am', 'vs_def': 'vs'}, inplace = True)
display(cars_data)
```

```python
# 3.1. Bifurcate 'cars_data' into 2 Subsets: 'cars_data_automatic' & 'cars_data_manual' using 'am' Column, each Subset having the following Columns: {am_def, mpg, disp, hp, wt}. 
```

```python
cars_data_automatic = cars_data[cars_data['am'] == 'Automatic']
cars_data_automatic [['am','mpg','disp','hp','wt']]
```

```python
cars_data_manual = cars_data[cars_data['am']=='Manual']
cars_data_manual [['am','mpg','disp','hp','wt']]
```

```python
# 3.2. Append both the Subsets 'cars_data_automatic' & 'cars_data_manual' to Create a New Dataset 'cars_data_drivetype_appended'.
```

```python
cars_data_drivetype_appended = cars_data_automatic.append (cars_data_manual)
cars_data_drivetype_appended
```

```python
# 4.1. Bifurcate 'cars_data' into 2 Subsets: 'cars_data_vshaped' & 'cars_data_straight' using 'vs_def' Column, such that Subset 'cars_data_vshaped' have the Columns: {am_def, mpg, disp} and Subset 'cars_data_straight' have the Columns: {am_def, hp, wt}.
```

```python
cars_data_vshaped = cars_data[cars_data['vs']=='Vshaped']
cars_data_vshaped [['am','mpg','disp','vs']]
```

```python
cars_data_straight = cars_data[cars_data['vs']=='Straight']
cars_data_straight[['am','hp','wt','vs']]
```

```python
# 4.2. Merge (Inner) both the Subsets 'cars_data_vshaped' & 'cars_data_straight' on 'am_def' column to Create a New Dataset 'cars_data_enginetype_merged'.
```

```python
import pandas as pd
cars_data_enginetype_merged = pd.merge(cars_data_vshaped, cars_data_straight, on='am', how='inner')
cars_data_enginetype_merged
```
