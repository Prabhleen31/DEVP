```python
# Descriptive Statistics of Categorical & Non-Categorical Variables
cars_data_noncat = cars_data[['mpg', 'disp', 'hp', 'drat', 'wt', 'qsec']] # Set of Non-Categorical Variables
cars_data_noncat
```

```python
cars_data_cat = cars_data[['cyl', 'vs', 'am', 'gear', 'carb']]  # Set of Categorical Variables

cars_data_cat
```

```python
import statsmodels.api
cars_data_noncat_stats = statsmodels.stats.descriptivestats.Description(cars_data_noncat).frame
cars_data_noncat_stats
```

```python
import statsmodels.api
cars_data_cat_stats = statsmodels.stats.descriptivestats.Description(cars_data_cat).frame
cars_data_cat_stats
```

```python
import pandas as pd
cyl_stats = pd.concat([cars_data['cyl'].value_counts(), cars_data['cyl'].value_counts(normalize=True).mul(100).round(0)], axis=1, keys=('count','percentage'))
cyl_stats
```

```python
vs_stats = pd.concat([cars_data['vs'].value_counts(), cars_data['vs'].value_counts(normalize=True).mul(100).round(0)], axis=1, keys=('count','percentage'))
vs_stats
```

```python
am_stats = pd.concat([cars_data['am'].value_counts(), cars_data['am'].value_counts(normalize=True).mul(100).round(0)], axis=1, keys=('count','percentage'))
am_stats
```
